```python
import pandas as pd
import numpy as np
import datetime
from datetime import datetime as dt
from datetime import timedelta 
import pytz
from sklearn.metrics import classification_report
from statsmodels.tsa.seasonal import STL
from statsmodels.tsa.seasonal import seasonal_decompose
from matplotlib import pyplot
from pycaret.regression import *
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("src/data/benchmarks.csv")
df["PayDate"] = pd.to_datetime(df["PayDate"], errors="coerce", utc=True)
df_temp = df.dropna().drop(columns = ["PatNum"])[(df["PayDate"] > (dt.now(tz=pytz.UTC) - timedelta(days=3650))) & (df["PayDate"] < (dt.now(tz=pytz.UTC) - timedelta(days=130)))]
df_temp = df_temp.groupby(pd.Grouper(key="PayDate", freq="1M")).sum()
#df_inputs = df_temp[df_temp.index.year == 2018]["PayAmt"].values
#df_inputs[0]
```

```python
result = seasonal_decompose(df_temp, model='additive')
result.plot()
pyplot.show()
```
